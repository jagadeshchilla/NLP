### Wordnet Lemmatizer

Lemmatization technique is like stemming. The output we will get after lemmatization is called lemma which is root word rather than root stem, the output stem. After lemmitization we will getting a valid word that means the same thing

NLTK provides WordNetLemmatizer class whuch is a thin Wrapper around the wordnet corpus. This class uses morphy() function to the WordNet CorpusReader class to find a lemma. Let us understand it with an example--


```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
```



as you can there is pos tag field in the function

```python
lemmatizer.lemmatize("going",pos="v")
```

```python
words=["eat","eating","eaten","eats","easily","fairly","fairness"]
```

```python
for word in words:
    print(word+"------>"+lemmatizer.lemmatize(word,pos="n"))
```

```python
for word in words:
    print(word+"------>"+lemmatizer.lemmatize(word,pos="v"))
```
